```python
# 直接执行shell命令
from invoke import run

cmd = "echo $PATH"
result = run(cmd, hide=True, warn=True)

print("执行状态：",result.ok)
print("返回结果：",result.stdout.splitlines()[-1])
```

```python
from invoke import tasks
```

```python
help(tasks)
```

```python
@task(default=True)
def hello(c):
    print("hello,world.")
```

```python
print(tasks.task(hello))
```

```python
tasks.call(hello,"hi")
```
